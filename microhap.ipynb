```python
def get_vcf_header(vcf_file_input):
    import gzip
    header_line = ""
    # set the opener if it is gzip
    opener = gzip.open if vcf_file_input.endswith(".gz") else open
    # open the file and find the header line
    with opener(vcf_file_input, "rt") as file:
        for line in file:
            if line.startswith("#CHROM"):
                header_line = line
                break
            elif not line.startswith("#"):
                break
    # format the header line to a list format
    if header_line:
        header_columns = header_line.strip("#").strip("\n").split("\t")
        return header_columns
    else:
        return "Unable to find the header, may be not a vcf file(if csv proceed & ignore)"

fixed_columns = get_vcf_header(vcf_file3)  # change file name here >>
print(f"These are the columns found :\n{fixed_columns}")
```

```python
# Convert the POS col to a list
snp_pos_list = df['POS'].sort_values().reset_index(drop=True).tolist()
snp_pos_list
```

### Clustering the Microhaplotypes

```python
"""TODO: Verify the logic in case of major error in the future"""
MAX_DISTANCE = 200  # set max distance
SNP_NUM = 3
snp_cluster_list = []
temp_cluster_list = [snp_pos_list[0]]

for i in range(1, len(snp_pos_list)):
    if (snp_pos_list[i] - snp_pos_list[i-1]) <= MAX_DISTANCE and len(temp_cluster_list) < SNP_NUM:
        temp_cluster_list.append(snp_pos_list[i])
    else:
        if len(temp_cluster_list) == SNP_NUM:
            snp_cluster_list.append(temp_cluster_list)
            # update the temp_cluster_list with the recent element
            temp_cluster_list = [snp_pos_list[i]]

# If at the end of the loop the size of temp_cluster_list is more than 3
if len(temp_cluster_list) > SNP_NUM:
    snp_cluster_list.append(temp_cluster_list)

# Display with small list chunks
#print(f"Cluster numbers: {snp_cluster_list}")
```

### Microhaplotypes Displayed

```python
i = 0
all_cluster_pos = [pos for cluster in snp_cluster_list for pos in cluster]
filt =  df['POS'].isin(all_cluster_pos)
df_save = df[filt]
df_save.head(10)
```
